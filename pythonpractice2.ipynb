```python
@smartdiv
def divison(a,b):
    return a/b
```

```python
def smartdiv(fun):
    def inner(a,b):
        if b==0:
            print('please enter non zero values in denominator')
        else:
            return fun(a,b)
    return inner
    
```

```python
divison(1,2)
```

```python
divison(1,0)
```
