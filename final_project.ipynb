below the function detect_face() function to detect the human faces in the photos 
in line 2 i convert the colour image to gray image and store the image in the gray
in line 3 use the haarcascade classifier to detect the faces in the photo haarcascade to find the positive images like face eyes in the image
line 4 detect multiscale to detect the faces more than 1 in the image here scalefactor is 1.3 and minneighbours if minneighbours is less than 5 it detect more false faces 
in line 5 if the length of the faces is zero it means the detector detect null it returns the null value
in line 7 x,y,w,h are the co-ordinates to find the face height and width x,and y co-ordinates
the function return the face from the image

```python
def detect_face(img):
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
    faces = face_cascade.detectMultiScale(gray, scaleFactor=1.3, minNeighbors=5);
    if (len(faces) == 0):
        return None, None
    (x, y, w, h) = faces[0]
    return gray[y:y+w, x:x+h], faces[0]
    
```

```python
def prepare_training_data(data_folder_path):
    dirs = os.listdir(data_folder_path)
    faces = []
    labels = []
    for dir_name in dirs:
        if not dir_name.startswith("s"):
            continue;
        label = int(dir_name.replace("s", ""))
        subject_dir_path = data_folder_path + "/" + dir_name
        subject_images_names = os.listdir(subject_dir_path)
        for image_name in subject_images_names:
            if image_name.startswith("."):
                continue;
            image_path = subject_dir_path + "/" + image_name
            image = cv2.imread(image_path)
            cv2.imshow("Training on image...", image)
            cv2.waitKey(100)
            face, rect = detect_face(image)
            if face is not None:
                faces.append(face)
                labels.append(label)
                cv2.destroyAllWindows()
                cv2.waitKey(1)
                cv2.destroyAllWindows()
    return faces, labels
                
```

here we use the lbph face recog nizer to train the data 
lbph facerecognizer is the inbuilt library of the open cv here it take the train images and find the centre pixel value and map the neighbor pixel values to binary values 0 and 1 if the neighbour value is graterthan or equal to centre pixel it map to 1 and if the neighbour value less than the centre value it map to 0 and convert the matrix into array
in line 1 we create an face recognizer 
in line to give the faces and labels lists to train the recognizer

```python
face_recognizer = cv2.face.LBPHFaceRecognizer_create()
face_recognizer.train(faces, np.array(labels))
```

here we store the face recognizer as the yml file we can re use the model without training data

draw_rectangle() function to draw the rectangle arround the face
here x,y,w,h coordinates are the height,width and x and y co ordinates and (0,255,0 ) color of the rectangle and 2 is the thicknes of the rectangle frame

```python
def draw_rectangle(img, rect):
    (x, y, w, h) = rect
    cv2.rectangle(img, (x, y), (x+w, y+h), (0, 255, 0), 2)
```

draw the test around the rectsngle frame 


```python
def draw_text(img, text, x, y):
    cv2.putText(img, text, (x, y), cv2.FONT_HERSHEY_PLAIN, 1.5, (0, 255, 0), 2)
```

here we write another function to predict the image 
in line 2 copy the imge we want to predict 
in line 3  to find the face of the image by fed the image into detect_face() function 
in line 4 store the predicted image into label variable
in line 5 store the text we have to draw top of the rectangle frame to label_text variable
in line 6 draw_rectangle() function to draw the rectangle around the face of the image
in line 7 draw the predicted label around the predicted image


```python
def predict(test_img):
    img = test_img.copy()
    face, rect = detect_face(img) 
    label= face_recognizer.predict(face)
    #print("confidence",confidence)
    label_text = subjects[np.array(label[0])]
    draw_rectangle(img, rect)
    draw_text(img, label_text, rect[0], rect[1]-5)
    return img
```
